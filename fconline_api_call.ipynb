#### ㄴ 매치 ID 리스트 생성
- 매치 상세 정보를 조회하는 API에 연동하기 위해 매치 ID 리스트 생성

```python
# 매치 기록 조회(최근 매치 ID 리스트 생성)
urlString = "https://open.api.nexon.com/fconline/v1/user/match?ouid=56f931578b228109e8e44599da41ea64&matchtype=50&offset=0&limit=100"
response = requests.get(urlString, headers = headers)

# 매치 ID 리스트 생성
match_id_list = response.json()
match_id_list
```

```python
# 내 경기 수
len(match_id_list)
```

#### ㄴ 매치 ID로 매치 상세 기록 조회

```python
# 매치 상세 기록 조회
match_info_list = []
for match_id in match_id_list:
    urlString = f"https://open.api.nexon.com/fconline/v1/match-detail?matchid={match_id}"

    res = requests.get(urlString, headers = headers)
    match_info_list.append(res.json())

# 예시) 매치ID별 정보
match_info_list[0]
```

#### ㄴ 데이터 프레임 생성
- json 파일로 API 호출
- json파일 기공하여 데이터프레임 생성
- 골 관련 정보도 있기 때문에 ROW의 기준은 골이 된다.

```python
# 필요한 데이터만 정리
columns = [
    'matchId','matchDate', 'matchType',"ouid","nickname", "matchResult", "matchEndType","foul", "injury", 
    "redCards", "yellowCards","possession", "averageRating", "dribble", "cornerKick", "offsideCount", 
    "shootTotal", "effectiveShootTotal", "goalTotal", "shootPenaltyKick", "goalPenaltyKick", 
    "passTry", "passSuccess", "shortPassTry", "shortPassSuccess", "longPassTry", "longPassSuccess", 
    "throughPassTry", "throughPassSuccess", "tackleTry", "tackleSuccess", "blockTry", "blockSuccess", 
    "goalTime", "x", "y", "type", "result", "spId", "assist", "assistSpId", "hitPost", "inPenalty"
]

match_df = pd.DataFrame(columns=columns)

# 주어진 데이터
for info in match_info_list:
    if 'error' in info.keys():
        pass
    else:
        match_info = info['matchInfo']

        data_list = []

        for player in match_info:
            match_detail = player['matchDetail']
            shoot = player['shoot']
            passing = player['pass']
            defence = player['defence']
            
            for shot in player.get('shootDetail', []):

                row = {
                    'matchId': info['matchId'],
                    'matchDate': info['matchDate'],
                    'matchType': info['matchType'],
                    "ouid": player['ouid'],
                    "nickname": player['nickname'],
                    "matchResult": match_detail['matchResult'],
                    "matchEndType": match_detail['matchEndType'],
                    "foul": match_detail['foul'],
                    "injury": match_detail['injury'],
                    "redCards": match_detail['redCards'],
                    "yellowCards": match_detail['yellowCards'],
                    "possession": match_detail['possession'],
                    "averageRating": match_detail['averageRating'],
                    "dribble": match_detail['dribble'],
                    "cornerKick": match_detail['cornerKick'],
                    "offsideCount": match_detail['offsideCount'],
                    "shootTotal": shoot['shootTotal'],
                    "effectiveShootTotal": shoot['effectiveShootTotal'],
                    "goalTotal": shoot['goalTotal'],
                    "shootPenaltyKick": shoot['shootPenaltyKick'],
                    "goalPenaltyKick": shoot['goalPenaltyKick'],
                    "passTry": passing['passTry'],
                    "passSuccess": passing['passSuccess'],
                    "shortPassTry": passing['shortPassTry'],
                    "shortPassSuccess": passing['shortPassSuccess'],
                    "longPassTry": passing['longPassTry'],
                    "longPassSuccess": passing['longPassSuccess'],
                    "throughPassTry": passing['throughPassTry'],
                    "throughPassSuccess": passing['throughPassSuccess'],
                    "tackleTry": defence['tackleTry'],
                    "tackleSuccess": defence['tackleSuccess'],
                    "blockTry": defence['blockTry'],
                    "blockSuccess": defence['blockSuccess'],
                    "goalTime": shot['goalTime'],
                    "x": shot['x'],
                    "y": shot['y'],
                    "type": shot['type'],
                    "result": shot['result'],
                    "spId": shot['spId'],
                    "assist": shot['assist'],
                    "assistSpId": shot['assistSpId'],
                    "hitPost": shot['hitPost'],
                    "inPenalty": shot['inPenalty']
                }
                
                data_list.append(row)

        # Pandas DataFrame 생성
        tmp_df = pd.DataFrame(data_list, columns=columns)
        
        match_df = pd.concat([match_df,tmp_df],axis=0, ignore_index=True)


# 데이터 출력
match_df.head()
```

```python
match_df.T
```

```python
# 내 경기 데이터 내보내기
match_df.to_csv("./data/my_match_data.csv", encoding='utf-8-sig', index=False)
```
